```python
import random

def hangman():
    # Word list and other game parameters
    word_list = ["python", "java", "javascript", "ruby", "swift"]
    word = random.choice(word_list)
    guessed_word = ['_'] * len(word)  # Initialize the word as underscores
    max_guesses = 6
    incorrect_guesses = 0
    guessed_letters = []

    print("Welcome to Hangman!")
    print("You have to guess the word. You can make a maximum of 6 incorrect guesses.")
    print("The word to guess has", len(word), "letters.")
    print(" ".join(guessed_word))

    # Game loop
    while incorrect_guesses < max_guesses:
        guess = input("\nGuess a letter: ").lower()

        # Check if input is a valid single letter
        if len(guess) != 1 or not guess.isalpha():
            print("Please enter a single letter.")
            continue

        # Check if the letter was already guessed
        if guess in guessed_letters:
            print(f"You already guessed the letter '{guess}'. Try a different one.")
            continue

        guessed_letters.append(guess)

        # Check if the guessed letter is in the word
        if guess in word:
            # Update guessed_word with the correct letter
            for i in range(len(word)):
                if word[i] == guess:
                    guessed_word[i] = guess
            print("Good guess!")
        else:
            incorrect_guesses += 1
            print(f"Incorrect guess! You have {max_guesses - incorrect_guesses} guesses left.")

        # Display the current state of the guessed word
        print(" ".join(guessed_word))

        # Check if the player has won
        if "_" not in guessed_word:
            print(f"Congratulations, you guessed the word '{word}'!")
            break
    else:
        print(f"Sorry, you've lost. The word was '{word}'.")

# Start the game
hangman()
```
